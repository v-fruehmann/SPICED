```python
# loop for 100 times
for fizzbuzz in range(1,101):
    # number divisible by both 3 & 5
    if fizzbuzz % 3 == 0 and fizzbuzz % 5 == 0:
        print("fizzbuzz")
        continue
    # number divisible by 3
    elif fizzbuzz % 3 == 0:
        print("fizz")
        continue
    # number divisible by 5
    elif fizzbuzz % 5 == 0:
        print("buzz")
        continue
    print(fizzbuzz)
```

```python
def fizzBuzz(n):
    for x in range(1, n + 1):
        if x % 3 == 0 and x % 5 == 0:
            print("fizzbuzz")
        elif x % 3 == 0:
            print("fizz")
        elif x % 5 == 0:
            print("buzz")
        else:
            print(x)
print(fizzBuzz(100))
```

```python
print(fizzBuzz(1000))
```
