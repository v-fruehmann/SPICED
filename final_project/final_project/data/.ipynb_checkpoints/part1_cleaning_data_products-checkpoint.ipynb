```python
def column_info(column):
    print('Column information {}'.format(column))
    print(f'Missing values: {df[column].isnull().sum()}')
    print('Column destribution:', pd.DataFrame(df[column].value_counts()))
    print(df[column].describe())
    if df[column].dtype == 'int64' or df[column].dtype == 'float64':
        median = df[column].median()
        perc25 = df[column].quantile(0.25)
        perc75 = df[column].quantile(0.75)
        IQR = perc75 - perc25
        print(f'Q1: {perc25}, Q3: {perc75}, IQR: {IQR} Outliers in the range of: [{perc25 - 1.5*IQR},{perc75 + 1.5*IQR}]')
        df[column].hist(bins = 10, label = 'IQR')
        plt.legend()
    print('\n')
```
