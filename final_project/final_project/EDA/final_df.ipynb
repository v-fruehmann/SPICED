## Reviews.csv

```python
df2 = pd.read_csv('reviews_full.csv')
df2
```

```python
df2.rename(columns={'text': 'review_text'}, inplace=True)
df2
```

```python
pos = df2[df2['negative or positive']=='positive'].astype(str)
pos = pos.groupby('asin')['review_text'].sum().reset_index()
pos.columns=['asin','good_reviews']

neg = df2[df2['negative or positive']=='negative'].astype(str)
neg = neg.groupby('asin')['review_text'].sum().reset_index()

neg.columns=['asin','bad_reviews']
df3 = pos.merge(neg,on='asin',how='outer')
df3.head(30)
```

```python
final_df = df.merge(df3,on='asin',how='outer')
final_df
```

```python
final_df.isnull().sum().sort_values(ascending=False)
```

```python
final_df["bad_reviews"].fillna("kein Kommentar", inplace = True)
```

```python
final_df["good_reviews"].fillna("kein Kommentar", inplace = True)
```

```python
final_df.to_csv('amazon_items_full.csv',index=False)
```
